```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#dropping all columns and rows with more than 70% NA Values
print('Shape Before Dropping', df.shape)
df.dropna(axis=1, thresh=int(0.3 * df.shape[0]), inplace=True)
df.dropna(axis=0, thresh=int(0.3 * df.shape[1]), inplace=True)
print('Shape After Dropping', df.shape)
```

```python
#printing all columns with na values, with their corresponding datatype
cols_with_na = df.columns[(1 - df.count()/df.shape[0]) > 0]
df[cols_with_na].info()
```

```python
#filling the missing values of each feature with its respective mean
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
df[cols_with_na] = imp.fit_transform(df[cols_with_na])
```

```python
#printing all columns with na values again, with their corresponding datatype, this should give an empty result
cols_with_na = df.columns[(1 - df.count()/df.shape[0]) > 0] 
df[cols_with_na].info()
```

```python
needed_date_features = ['year', 'month', 'day']
for item in needed_date_features:
    df[item] = getattr(df['Date'].dt, item)
```

```python
target_features = ['target', 'target_min','target_max', 'target_variance', 'target_count']
to_be_removed = ['Place_ID X Date', 'Date', 'Place_ID']
working_df = df.copy()
```

```python
X = working_df.drop(target_features + to_be_removed, axis=1)
y = working_df['target']
```

```python
predicting_features = X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
#Standardizing the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from lightgbm import LGBMRegressor
```

```python
model = LGBMRegressor(random_state=100)
model.fit(X_train, y_train)
prediction = model.predict(X_test)
prediction_df = pd.DataFrame(data=np.hstack([y_test.values.reshape(-1,1), prediction.reshape(-1,1)]), columns=["Actual Value", "Predicted Value"])
prediction_df.head(10)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("MSE Score", mean_squared_error(y_test,prediction))
print("RMSE Score", mean_squared_error(y_test,prediction)**0.5)
```
